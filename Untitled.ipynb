```python
#1.Adding date features

# Import training and test data
train = pd.read_csv('trainv.csv')
test = pd.read_csv('testv.csv')

# DATES FEATURES
def date_features(df):
    # Date Features
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df.date.dt.year
    df['month'] = df.date.dt.month
    df['day'] = df.date.dt.day
    df['dayofyear'] = df.date.dt.dayofyear
    df['dayofweek'] = df.date.dt.dayofweek
    df['weekofyear'] = df.date.dt.weekofyear
    
    # Additionnal Data Features
    df['day^year'] = np.log((np.log(df['dayofyear'] + 1)) ** (df['year'] - 2000))
    
    # Drop date
    df.drop('date', axis=1, inplace=True)
    
    return df

# Dates Features for Train, Test
train, test = date_features(train), date_features(test)
```

```python
# Daily Average, Monthly Average for train
train['daily_avg']  = train.groupby(['item','hospital','dayofweek'])['qty'].transform('mean')
train['monthly_avg'] = train.groupby(['item','hospital','month'])['qty'].transform('mean')
train = train.dropna()

# Average quantity for Day_of_week = d per Item,hospital
daymonth_avg = train.groupby(['item','hospital','dayofweek'])['qty'].mean().reset_index()
# Average quantity for Month = m per Item,hospital
monthly_avg = train.groupby(['item','hospital','month'])['qty'].mean().reset_index()
#Average quantity for a day
daily_avg= train.groupby(['item','hospital','dayofweek'])['qty'].mean().reset_index()
```

```python
train[:5]
```

```python
# Merge Test with Daily Avg, Monthly Avg
def merge(df1, df2, col,col_name):
    
    df1 =pd.merge(df1, df2, how='left', on=None, left_on=col, right_on=col,
            left_index=False, right_index=False, sort=True,
             copy=True, indicator=False)
    
    df1 = df1.rename(columns={'qty':col_name})
    return df1

```

```python
# Add Daily_avg and Monthly_avg features to test 
test = merge(test, daily_avg,['item','hospital','dayofweek'],'daily_avg')
test = merge(test, monthly_avg,['item','hospital','month'],'monthly_avg')

# Quantity Rolling mean sequence per item (un weighted mean of last n values)
rolling_10 = train.groupby(['item'])['qty'].rolling(10).mean().reset_index().drop('level_1', axis=1)
train['rolling_mean'] = rolling_10['qty'] 

# 90 last days of training rolling mean sequence added to test data
rolling_last90 = train.groupby(['item','hospital'])['rolling_mean'].tail(90).copy()
test['rolling_mean'] = rolling_last90.reset_index().drop('index', axis=1)

# Shifting rolling mean 3 months
train['rolling_mean'] = train.groupby(['item'])['rolling_mean'].shift(90) # Create a feature with rolling mean of day - 90
train.head()
```

Let us keep the monthly & daily average since it has the highest correlation with qty; and remove other features highly correlated to each other.

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
# Clean features highly correlated to each others

for df in [train, test]:
    df.drop(['dayofyear', 
                  'weekofyear',
                  'daily_avg',
                  'day',
                  'month',
                  'item',
                  'hospital'],
                 axis=1, 
                 inplace=True)
    
# Features Scaling (except sales)
qty_vaccine, id_series = train['qty'], test['id']
# Features Scaling
train = (train - train.mean()) / train.std()
test = (test - test.mean()) / test.std()
# Retrieve actual Sales values and ID
train['qty'] = qty_vaccine
test['id'] = id_series

# Training Data
X_train = train.drop('qty', axis=1).dropna()
y_train = train['qty']
# Test Data
test.sort_values(by=['id'], inplace=True)
X_test = test.drop('id', axis=1)
#df = train
df_train = train.copy()

# Train Test Split
X_train , X_test ,y_train, y_test = train_test_split(df_train.drop('qty',axis=1),df_train.pop('qty'), random_state=123, test_size=0.2)

# XGB Model
matrix_train = xgb.DMatrix(X_train, label = y_train)
matrix_test = xgb.DMatrix(X_test, label = y_test)

# Run XGB 
model = xgb.train(params={'objective':'reg:linear','eval_metric':'mae'}
                ,dtrain = matrix_train, num_boost_round = 500, 
                early_stopping_rounds = 20, evals = [(matrix_test,'test')],)
```

```python
#submission = pd.DataFrame(X_test.pop('id'))
y_test = model.predict(xgb.DMatrix(X_test), ntree_limit = model.best_ntree_limit)

#submission['qty']= y_test

#submission.to_csv('sub3.csv',index=False)
y_test
```
